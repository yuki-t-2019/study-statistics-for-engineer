# 第1章 統計学の基礎

<br>

## 1.1 統計学とは

統計学とは、データから現象の法則性を見つけるための手段や方法論に関する学問である。
1. 現象の法則性を知るために、すべてを丹念に調べ、規則性から法則を見出す
1. 一部を観察して、そこから論理性のある推測で全体の法則性を発見する

<br>

上記1は**記述統計学**と定義され、  
また、2は確率論をもとに1の上に打ち立てた方法論の体系であり、**統計的推測**といわれている。

統計学はこの2つの分野が合わさったものである。

<br>

全体(集団)の法則性を知るために全体を調べる考え方が**全数調査**といわれる考え方であり、  
代表的なものに国勢調査がある。

それに対して全体(**母集団**)から正しくサンプル(**標本**)を抜き出し、  
サンプルから法則性を見出し、確率論をもって全体を推測する方法を**標本による調査**と呼んでいる。

<br>

## 1.2 統計データと統計手法
データはその性質によっていくつかのタイプに分けられる。

<br>

### 1.2.1 量的データと質的データ
下の表1のように、データが定量的な値で与えられるものを**量的データ**と呼ぶ。  
＊長さ、重さ、退席、面積、金額、温度、時間など、数値でその値を測定できるもの

<center>
表1
</center>

| 学生  |   1   |   2   |   3   |
| :---: | :---: | :---: | :---: |
|  身長 |  165  |  168  |  152  |

一方、下の表2のように数値として観測できずに、あるカテゴリーに属しているorある状態にある、  
ということだけが分かるデータを**質的データ**と呼ぶ。  
＊性別、学歴、天気、地域など、数値でその値を側的できないもの

<center>
表2
</center>

| 学生  |   1   |   2   |   3   |
| :---: | :---: | :---: | :---: |
|  性別 |  男   |  女   |  男   |

<br>

### 1.2.2 1次元データと多次元データ

上の表1や表2などのように、1人の学生に対して1つの観測値だけ与えられているデータを**1次元データ**と呼ぶ。  

一方、下の表3のように1人の学生に対して2つの観測値が与えられているデータを**2次元データ**と呼ぶ。

<center>
表3
</center>

| 学生  |   1   |   2   |   3   |
| :---: | :---: | :---: | :---: |
| 身長  |  178  |  165  |  168  |
| 体重  |  63   |  62   |  69   |

統計学では2次元以上のデータを**多次元データ**と呼んでいる。

<br>

### 1.2.3 時系列データとクロスセクション・データ
下の表4は、昭和15年から昭和60年までの人口推移を示している。

<center>
表4
</center>

|     年次     |   15   |   20   |   …   |   60    |
| :----------: | :---:  | :---:  | :---: | :-----: |
| 人口[1000人] | 71,933 | 72,147 |   …   | 121,049 |

このような同一の対象の異なった時点での観測値からなるデータを**時系列データ**と呼ぶ。

<br>

一方、下の表5は、世界各国の1986年における推定人口である。

<center>
表5
</center>

|     年次     |   15   |   20   |   …   |   60    |
| :----------: | :---:  | :---:  | :---: | :-----: |
| 人口[1000人] | 71,933 | 72,147 |   …   | 121,049 |

このように、いくつかの異なった対象について調査・実験を行い、  
観測値を得た場合のデータを**クロスセクション・データ**と呼ぶ。
